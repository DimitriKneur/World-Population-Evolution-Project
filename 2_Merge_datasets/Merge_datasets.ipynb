```python
import pandas as pd
```

```python
df_birth_rate_PIB = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/Birth_Rate_PIB.csv")
display(df_birth_rate_PIB)

df_demography = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/Demography_dataset.csv")
display(df_demography)

df_idh = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/IDH_table.csv")
display(df_idh)

df_pollution = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/pollution.csv')
display(df_pollution)

df_employment = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/employment.csv")
display(df_employment)

df_urban = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/urban.csv")
display(df_urban)

df_death_rate = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_death_rate.csv", index_col=0)
display(df_death_rate)

df_pop_active = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_population_active.csv", index_col=0)
display(df_pop_active)

df_pop_65_and_above = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_65_above.csv", index_col=0)
display(df_pop_65_and_above)

df_cotisations_sociales = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_cot_sociale.csv", index_col=0)
display(df_cotisations_sociales)

df_life_expectancy = pd.read_csv("d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/Dataset_life_expectancy.csv", index_col=0)
display(df_life_expectancy)

df_gov_health_expenditure = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_gov_health_expenditure.csv', index_col=0)
display(df_gov_health_expenditure)

df_undernourishment = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_undernourishment.csv', index_col=0)
display(df_undernourishment)

df_water_access = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_water_access.csv', index_col=0)
display(df_water_access)

df_mortality_cancer_etc = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_mortality_cvd_crd_cancer_diabete_30_70.csv', index_col=0)
display(df_mortality_cancer_etc)

df_uhc_service_coverage = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/dataset_uhc_service_coverage.csv', index_col=0)
display(df_uhc_service_coverage)

df_pop_growth_rate = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/population_growth_rate.csv', index_col=0)
display(df_pop_growth_rate)
```

```python
df_employment.rename(columns={"Employement rate":"Female_employment_rate"}, inplace=True)
display(df_employment)
```

```python
df_pop_growth_rate.rename(columns={"economy":"Country Code"}, inplace=True)
display(df_pop_growth_rate)
```

```python
merged_final_dataset = df_birth_rate_PIB.merge(
                                                df_demography,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=["Country Name", "Country Code", "Year"],
                                                how="inner"
                                                ).merge(
                                                        df_idh,
                                                        left_on=["Country Name", "Country Code", "Year"],
                                                        right_on=["Entity", "Code", "Year"],
                                                        how="inner"
                                                        ).merge(
                                                                df_pollution,
                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                how="inner"
                                                                ).merge(
                                                                        df_employment,
                                                                        left_on=["Country Name", "Country Code", "Year"],
                                                                        right_on=["Country Name", "Country Code", "Year"],
                                                                        how="inner"
                                                                        ).merge(
                                                                                df_urban,
                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                how="inner"
                                                                                ).merge(
                                                                                        df_death_rate,
                                                                                        left_on=["Country Code", "Year"],
                                                                                        right_on=["Country Code", "Year"],
                                                                                        how="inner"
                                                                                        ).merge(
                                                                                                df_pop_active,
                                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                                how="inner"
                                                                                                ).merge(
                                                                                                        df_pop_65_and_above,
                                                                                                        left_on=["Country Name", "Country Code", "Year"],
                                                                                                        right_on=["Country Name", "Country Code", "Year"],
                                                                                                        how="inner"
                                                                                                        ).merge(
                                                                                                                df_cotisations_sociales,
                                                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                                                how="left"
                                                                                                                ).merge(
                                                                                                                        df_life_expectancy,
                                                                                                                        left_on=["Country Code", "Year"],
                                                                                                                        right_on=["Country Code", "Year"],
                                                                                                                        how="inner"
                                                                                                                        ).merge(
                                                                                                                                df_gov_health_expenditure,
                                                                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                                                                how="left"
                                                                                                                                ).merge(
                                                                                                                                        df_undernourishment,
                                                                                                                                        left_on=["Country Name", "Country Code", "Year"],
                                                                                                                                        right_on=["Country Name", "Country Code", "Year"],
                                                                                                                                        how="left"
                                                                                                                                        ).merge(
                                                                                                                                                df_water_access,
                                                                                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                how="left"
                                                                                                                                                ).merge(
                                                                                                                                                        df_mortality_cancer_etc,
                                                                                                                                                        left_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                        right_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                        how="left"
                                                                                                                                                        ).merge(
                                                                                                                                                                df_uhc_service_coverage,
                                                                                                                                                                left_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                                right_on=["Country Name", "Country Code", "Year"],
                                                                                                                                                                how="left"
                                                                                                                                                                ).merge(
                                                                                                                                                                        df_pop_growth_rate,
                                                                                                                                                                        left_on=["Country Code", "Year"],
                                                                                                                                                                        right_on=["Country Code", "Year"],
                                                                                                                                                                        how="inner"
                                                                                                                                                                        )
display(merged_final_dataset)
```

```python
joined_merged_final_dataset = merged_final_dataset.drop(['Unnamed: 0_x',
                                                        'Series Name',
                                                        'Series Code',
                                                        'Unnamed: 0_y',
                                                        'Entity',
                                                        'Code'
                                                        ], 
                                                        axis=1
                                                        )
display(joined_merged_final_dataset)
```
