```python
df = merged_dataset_to_clean.loc[:,['Country Name', 'Country Code', 'Year', 'Female_employment_rate']]
df
```

```python
# Drop 10 countries Andorra, Antigua and Barbuda, Dominica, Grenada, Kiribati, Liechtenstein, Marshall Islands, Palau, Seychelles, Tuvalu
list_to_drop = ['Andorra', 'Antigua and Barbuda', 'Dominica', 'Grenada', 'Kiribati', 'Liechtenstein', 'Marshall Islands', 'Palau', 'Seychelles', 'Tuvalu']
df = df.drop(df[df['Country Name'].isin(list_to_drop)].index)
```

```python
df[df['Female_employment_rate'].isna()].groupby('Country Name').count()
```

```python
df[df['Country Name']=='Afghanistan']
```

```python
# replace the Nan value by Afghanistan female % of labor force in 2021 (19.8) https://wdi.worldbank.org/table/2.2
df.loc[(df['Country Name'] == 'Afghanistan') & (df['Year'] == 2021), 'Female_employment_rate'] = 19.8
```

```python
df[df['Country Name']=='Myanmar']
```

```python
# replace the Nan value by Myanmar female Labor force participation rate in 2021 which is 40.99. https://www.indexmundi.com/facts/myanmar/indicator/SL.TLF.CACT.FE.ZS#:~:text=Labor%20force%20participation%20rate%2C%20female%20%28%25%20of%20female,while%20its%20lowest%20value%20was%2040.99%20in%202021.
df.loc[(df['Country Name'] == 'Myanmar') & (df['Year'] == 2021), 'Female_employment_rate'] = 40.99
```
