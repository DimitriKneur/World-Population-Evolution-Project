```python
import pandas as pd
```

```python
df_birth_rate_PIB = pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\Birth_Rate_PIB.csv", index_col=0)
display(df_birth_rate_PIB)

df_total_population = pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\Dataset_male_and_female_population.csv", index_col=0)
display(df_total_population)

df_death_rate = pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\dataset_death_rate.csv", index_col=0)
display(df_death_rate)

df_pop_65_and_above = pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\dataset_65_above.csv", index_col=0)
display(df_pop_65_and_above)

df_life_expectancy = pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\Dataset_life_expectancy.csv", index_col=0)
display(df_life_expectancy)

df_pop_growth_rate =pd.read_csv("C:\\Users\\celia\\OneDrive\\桌面\\BootCamp\\JEDHA\\fullstack\\projet\\Demoday\\1_Create_and_join_datasets\\population_growth_rate.csv", index_col=0)
display(df_pop_growth_rate)
```

```python
df_pop_growth_rate.rename(columns={"economy":"Country Code"}, inplace=True)
```

```python
df = df_birth_rate_PIB.merge(df_death_rate,
                                        left_on=["Country Code", "Year"],
                                        right_on=["Country Code", "Year"],
                                        how="inner"
                                ).merge(
                                        df_pop_65_and_above,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=[ "Country Name", "Country Code", "Year"],
                                                how="inner"
                                        ).merge(
                                                df_pop_growth_rate,
                                                        left_on=["Country Code", "Year"],
                                                        right_on=["Country Code", "Year"],
                                                        how="inner"
                                                ).merge(
                                                    df_total_population,
                                                        left_on=["Country Name", "Country Code", "Year"],
                                                        right_on=["Country Name", "Country Code", "Year"],
                                                        how="inner"
                                                ).merge(
                                                    df_life_expectancy,
                                                        left_on=[ "Country Code", "Year"],
                                                        right_on=[ "Country Code", "Year"],
                                                        how="inner"

                                                )
display(df)
```
