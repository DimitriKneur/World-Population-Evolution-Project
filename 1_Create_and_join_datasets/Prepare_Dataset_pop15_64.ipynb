```python
import pandas as pd
```

```python
dataset_population_active = pd.read_csv('population_15_64.csv')
display(dataset_population_active)
```

```python
dataset_population_active_unpivot = pd.melt(
                                        dataset_population_active,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="Population_15_64_rate"    
                                                                )
display(dataset_population_active_unpivot)
```

```python
dataset_population_active = dataset_population_active_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_population_active)
```

```python
dataset_population_active.to_csv('dataset_population_active.csv')
```
