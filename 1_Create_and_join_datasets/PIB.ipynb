Unpivoter les colonnes Years

```python
df_melted = pd.melt(df_PIB, id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], var_name='Year', value_name='GDP per Capita')
df_melted.head()
```

```python
df_melted.tail(10)
```

Supprission des neufs derniÃ¨res lignes

```python
df_PIB_tronque = df_melted.iloc[:-7]
df_PIB_tronque.tail()
```

Modifier le format des valeurs de la colonne "Year"

```python
df_PIB_tronque["Year"] = df_PIB_tronque["Year"].apply(lambda x: x[:4]) 
display(df_PIB_tronque)
```

Affichage du deuxiÃ¨me dataframe (df_Birth_Rate)

```python
df_Birth_Rate = pd.read_csv("total_birth_rate.csv", encoding="unicode_escape")
df_Birth_Rate.head()
```

Unpivoter les colonnes Years

```python
df_Birth_Rate_unpivot = pd.melt(df_Birth_Rate, id_vars=['economy'], var_name='Year', value_name='Birth Rate')
df_Birth_Rate_unpivot.head()
```

Renommer la colonne "economy" Ã  "Country Name"

```python
df_Birth_Rate_unpivot.rename(columns={'economy': 'Country Code'}, inplace=True)
df_Birth_Rate_unpivot.head()
```

Modifier le format des valeurs de la colonne "Year"

```python
df_Birth_Rate_unpivot["Year"] = [yearacronym.replace("YR", "") for yearacronym in df_Birth_Rate_unpivot["Year"]]
display(df_Birth_Rate_unpivot) 
```

Merger les deux Dataframes (df_Birth_Rate_unpivot et df_PIB_tronque)

```python
df_Birth_Rate_PIB = pd.merge(df_Birth_Rate_unpivot, df_PIB_tronque, on=['Country Code', 'Year'])
display(df_Birth_Rate_PIB)
```

Modifier l'emplacement des colonnes

```python
colonnes = df_Birth_Rate_PIB.columns.tolist()
colonnes.insert(0, colonnes.pop(3))
df_Birth_Rate_PIB = df_Birth_Rate_PIB[colonnes]
display(df_Birth_Rate_PIB)
```

Exporter le dataframe "df_Birth_Rate_PIB" sous format CSV

```python
df_Birth_Rate_PIB.to_csv("Birth_Rate_PIB.csv")
```
