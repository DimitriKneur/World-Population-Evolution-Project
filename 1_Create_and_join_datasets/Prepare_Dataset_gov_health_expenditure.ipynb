```python
import pandas as pd
```

```python
dataset_gov_health_exp = pd.read_csv('gov_health_expenditure.csv')
display(dataset_gov_health_exp)
```

```python
dataset_gov_health_exp_unpivot = pd.melt(
                                        dataset_gov_health_exp,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="gov_health_exp_as_perc_of_gdp"    
                                                                )
display(dataset_gov_health_exp_unpivot)
```

```python
dataset_gov_health_exp_unpivot = dataset_gov_health_exp_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_gov_health_exp_unpivot)
```

```python
dataset_gov_health_exp_unpivot.to_csv('dataset_gov_health_expenditure.csv')
```
