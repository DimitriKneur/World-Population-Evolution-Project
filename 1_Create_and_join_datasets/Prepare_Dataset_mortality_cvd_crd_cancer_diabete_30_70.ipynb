```python
import pandas as pd
```

```python
dataset_mortality_cancer_etc = pd.read_csv('mortality_cvd_crd_cancer_diabete_30_70.csv')
display(dataset_mortality_cancer_etc)
```

```python
dataset_mortality_cancer_etc_unpivot = pd.melt(
                                        dataset_mortality_cancer_etc,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="mortality_cvd_crd_cancer_diabete_30_70"    
                                                                )
display(dataset_mortality_cancer_etc_unpivot)
```

```python
dataset_mortality_cancer_etc_unpivot = dataset_mortality_cancer_etc_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_mortality_cancer_etc_unpivot)
```

```python
dataset_mortality_cancer_etc_unpivot.to_csv('dataset_mortality_cvd_crd_cancer_diabete_30_70.csv')
```
