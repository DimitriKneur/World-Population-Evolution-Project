```python
import pandas as pd
```

```python
dataset_water_access = pd.read_csv('water_access.csv')
display(dataset_water_access)
```

```python
dataset_water_access_unpivot = pd.melt(
                                        dataset_water_access,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="population_rate_with_water_access"    
                                                                )
display(dataset_water_access_unpivot)
```

```python
dataset_water_access_unpivot = dataset_water_access_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_water_access_unpivot)
```

```python
dataset_water_access_unpivot.to_csv('dataset_water_access.csv')
```
