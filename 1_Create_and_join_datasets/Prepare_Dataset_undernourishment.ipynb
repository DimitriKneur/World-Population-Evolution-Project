```python
import pandas as pd
```

```python
dataset_undernourishment = pd.read_csv('undernourishment.csv')
display(dataset_undernourishment)
```

```python
dataset_undernourishment_unpivot = pd.melt(
                                        dataset_undernourishment,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="undernourished_population_rate"    
                                                                )
display(dataset_undernourishment_unpivot)
```

```python
dataset_undernourishment_unpivot = dataset_undernourishment_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_undernourishment_unpivot)
```

```python
dataset_undernourishment_unpivot.to_csv('dataset_undernourishment.csv')
```
