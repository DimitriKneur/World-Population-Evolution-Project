```python
import pandas as pd
```

```python
dataset_cot_sociale = pd.read_csv('taux_cotisations_sociales_par_pays.csv')
display(dataset_cot_sociale)
```

```python
dataset_cot_sociale = pd.melt(
                                        dataset_cot_sociale,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="Cotisation sociale"    
                                                                )
display(dataset_cot_sociale)
```

```python
dataset_cot_sociale = dataset_cot_sociale.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_cot_sociale)
```

```python
dataset_cot_sociale.to_csv('dataset_cot_sociale.csv')
```
