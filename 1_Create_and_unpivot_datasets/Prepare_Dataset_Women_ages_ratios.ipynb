```python
import pandas as pd
```

```python
dataset_women_age_ratios = pd.read_csv("d:/Formations/Data_Analysis/Data_Analyst_Fullstack/Projects/Demo_day/0_Source_Datasets/Female_population_between_19_and_44_statistics_2000_2021.csv")
display(dataset_women_age_ratios)
```

```python
list(dataset_women_age_ratios["Series Name"].unique())
```

```python
dataset_women_15_19 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 15-19, female (% of female population)"]
display(dataset_women_15_19)
dataset_women_20_24 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 20-24, female (% of female population)"]
display(dataset_women_20_24)
dataset_women_25_29 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 25-29, female (% of female population)"]
display(dataset_women_25_29)
dataset_women_30_34 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 30-34, female (% of female population)"]
display(dataset_women_30_34)
dataset_women_35_39 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 35-39, female (% of female population)"]
display(dataset_women_35_39)
dataset_women_40_44 = dataset_women_age_ratios[dataset_women_age_ratios['Series Name'] == "Population ages 40-44, female (% of female population)"]
display(dataset_women_40_44)
```

```python
dataset_women_15_19_unpivot = pd.melt(
                                                        dataset_women_15_19,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="15_to_19_women_rate"    
                                                                )
display(dataset_women_15_19_unpivot)

dataset_women_20_24_unpivot = pd.melt(
                                                        dataset_women_20_24,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="20_to_24_women_rate"    
                                                                )
display(dataset_women_20_24_unpivot)

dataset_women_25_29_unpivot = pd.melt(
                                                        dataset_women_25_29,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="25_to_29_women_rate"    
                                                                )
display(dataset_women_25_29_unpivot)

dataset_women_30_34_unpivot = pd.melt(
                                                        dataset_women_30_34,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="30_to_34_women_rate"    
                                                                )
display(dataset_women_30_34_unpivot)

dataset_women_35_39_unpivot = pd.melt(
                                                        dataset_women_35_39,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="35_to_39_women_rate"    
                                                                )
display(dataset_women_35_39_unpivot)

dataset_women_40_44_unpivot = pd.melt(
                                                        dataset_women_40_44,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="40_to_44_women_rate"    
                                                                )
display(dataset_women_40_44_unpivot)
```

```python
dataset_women_15_19_unpivot["Year"] = dataset_women_15_19_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_15_19_unpivot)
dataset_women_20_24_unpivot["Year"] = dataset_women_20_24_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_20_24_unpivot)
dataset_women_25_29_unpivot["Year"] = dataset_women_25_29_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_25_29_unpivot)
dataset_women_30_34_unpivot["Year"] = dataset_women_30_34_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_30_34_unpivot)
dataset_women_35_39_unpivot["Year"] = dataset_women_35_39_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_35_39_unpivot)
dataset_women_40_44_unpivot["Year"] = dataset_women_40_44_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_women_40_44_unpivot)
```

```python
final_dataset_women_15_19 = dataset_women_15_19_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_15_19)
final_dataset_women_20_24 = dataset_women_20_24_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_20_24)
final_dataset_women_25_29 = dataset_women_25_29_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_25_29)
final_dataset_women_30_34 = dataset_women_30_34_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_30_34)
final_dataset_women_35_39 = dataset_women_35_39_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_35_39)
final_dataset_women_40_44 = dataset_women_40_44_unpivot.drop(['Series Name','Series Code'], axis=1)
display(final_dataset_women_40_44)
```

```python
final_dataset_women_age_ratios = final_dataset_women_15_19.merge(
                                                                 final_dataset_women_20_24,
                                                                 left_on=['Country Name', 'Country Code', 'Year'],
                                                                 right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                                ).merge(
                                                                        final_dataset_women_25_29,
                                                                        left_on=['Country Name', 'Country Code', 'Year'],
                                                                        right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                                       ).merge(
                                                                               final_dataset_women_30_34,
                                                                               left_on=['Country Name', 'Country Code', 'Year'],
                                                                               right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                                              ).merge(
                                                                                      final_dataset_women_35_39,
                                                                                      left_on=['Country Name', 'Country Code', 'Year'],
                                                                                      right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                                                     ).merge(
                                                                                             final_dataset_women_40_44,
                                                                                             left_on=['Country Name', 'Country Code', 'Year'],
                                                                                             right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                                                            )

display(final_dataset_women_age_ratios)
```

```python
final_dataset_women_age_ratios.to_csv("Dataset_women_age_ratios.csv")
```
