```python
import pandas as pd
```

```python
dataset_death_rate = pd.read_csv('total_death_rate.csv')
display(dataset_death_rate)
```

```python
dataset_death_rate_unpivot = pd.melt(
                                        dataset_death_rate,
                                        id_vars = "economy",
                                        var_name = "Year",
                                        value_name="Death_Rate"    
                                                                )
display(dataset_death_rate_unpivot)
```

```python
dataset_death_rate_unpivot.rename(columns={"economy": "Country Code"}, inplace=True)

display(dataset_death_rate_unpivot)
```

```python
dataset_death_rate_unpivot["Year"] = dataset_death_rate_unpivot["Year"].apply(lambda x : x[2:])
display(dataset_death_rate_unpivot)
```

```python
dataset_death_rate_unpivot.to_csv('dataset_death_rate.csv')
```
