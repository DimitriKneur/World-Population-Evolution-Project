```python
dataset_male_female_population = pd.read_csv("D:/Formations/Data_Analysis/Data_Analyst_Fullstack/Projects/Demo_day/0_Source_Datasets/Population_male_female_total.csv")
display(dataset_male_female_population)
```

```python
dataset_male_female_population_justmale = dataset_male_female_population[dataset_male_female_population['Series Name'] == 'Population, male']
display(dataset_male_female_population_justmale)
```

```python
dataset_male_female_population_justmale_unpivot = pd.melt(
                                                        dataset_male_female_population_justmale,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="Population_male"    
                                                                )
display(dataset_male_female_population_justmale_unpivot)
```

```python
dataset_male_female_population_justmale_unpivot["Year"] = dataset_male_female_population_justmale_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_male_female_population_justmale_unpivot)
```

```python
dataset_male_female_population_justmale_unpivot.columns
```

```python
dataset_male_population = dataset_male_female_population_justmale_unpivot.drop(['Series Name','Series Code'], axis=1)
dataset_male_population
```

```python
dataset_male_female_population_just_female = dataset_male_female_population[dataset_male_female_population['Series Name'] == 'Population, female']
display(dataset_male_female_population_just_female)
```

```python
dataset_male_female_population_just_female_unpivot = pd.melt(
                                                        dataset_male_female_population_just_female,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="Population_female"    
                                                                )
display(dataset_male_female_population_just_female_unpivot)
```

```python
dataset_male_female_population_just_female_unpivot["Year"] = dataset_male_female_population_just_female_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_male_female_population_just_female_unpivot)
```

```python
dataset_female_population = dataset_male_female_population_just_female_unpivot.drop(['Series Name','Series Code'], axis=1)
dataset_female_population
```

```python
dataset_male_female_population_just_total = dataset_male_female_population[dataset_male_female_population['Series Name'] == 'Population, total']
display(dataset_male_female_population_just_total)
```

```python
dataset_male_female_population_just_total_unpivot = pd.melt(
                                                        dataset_male_female_population_just_total,
                                                        id_vars = ["Country Name","Country Code","Series Name","Series Code"],
                                                        var_name = "Year",
                                                        value_name="Population_total"    
                                                                )
display(dataset_male_female_population_just_total_unpivot)
```

```python
dataset_male_female_population_just_total_unpivot["Year"] = dataset_male_female_population_just_total_unpivot["Year"].apply(lambda x : x[:4])
display(dataset_male_female_population_just_total_unpivot)
```

```python
dataset_total_population = dataset_male_female_population_just_total_unpivot.drop(['Series Name','Series Code'], axis=1)
display(dataset_total_population)
```

```python
# Outer join: we keep all the ids we find in both tables and fill the missing values in with NaN
df_all_population = dataset_male_population.merge(
                                         dataset_female_population,
                                         left_on=['Country Name', 'Country Code', 'Year'],
                                         right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                         ).merge(
                                                 dataset_total_population,
                                                 left_on=['Country Name', 'Country Code', 'Year'],
                                                 right_on=['Country Name', 'Country Code', 'Year'], how='inner'
                                                )
```

```python
display(df_all_population)
```

```python
df_all_population.info()
```

```python
def convert_to_integer(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return value

df_all_population["Year"] = df_all_population.Year.astype(int)
df_all_population["Population_male"] = df_all_population["Population_male"].apply(convert_to_integer)
df_all_population["Population_female"] = df_all_population["Population_female"].apply(convert_to_integer)
df_all_population["Population_total"] = df_all_population["Population_total"].apply(convert_to_integer)


df_all_population.info()
```
