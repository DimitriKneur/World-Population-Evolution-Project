```python
import pandas as pd
```

```python
df_population = pd.read_csv("Dataset_male_and_female_population.csv")
display(df_population)
df_median_age = pd.read_csv("Dataset_median_age.csv")
display(df_median_age)
df_women_age_ratios = pd.read_csv("Dataset_women_age_ratios.csv")
display(df_population)
```

```python
merged_demography_dataset = df_population.merge(
                                                df_women_age_ratios,
                                                left_on=["Country Name", "Country Code", "Year"],
                                                right_on=["Country Name", "Country Code", "Year"],
                                                how="inner"
                                                ).merge(
                                                        df_median_age,
                                                        left_on=["Country Code", "Year"],
                                                        right_on=["Country Code", "Year"],
                                                        how="inner"
                                                        )
display(merged_demography_dataset)
```

```python
final_merged_demography_dataset = merged_demography_dataset.drop(["Unnamed: 0_x", "Unnamed: 0_y", "Unnamed: 0"], axis=1)
display(final_merged_demography_dataset)
```

```python
final_merged_demography_dataset.to_csv('Demography_dataset.csv')
```
