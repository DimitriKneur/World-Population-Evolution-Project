```python
import pandas as pd
```

```python
dataset_65_above = pd.read_csv('population_65_and_above.csv')
display(dataset_65_above)
```

```python
dataset_65_above_unpivot = pd.melt(
                                        dataset_65_above,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="Population_65_above"    
                                                                )
display(dataset_65_above_unpivot)
```

```python
dataset_65_above = dataset_65_above_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_65_above)
```

```python
dataset_65_above.to_csv('dataset_65_above.csv')
```
