```python
df_birth_rate = pd.read_csv(io.BytesIO(uploaded['Table_principale (3).csv']),sep=";") # ";" pour afficher plusieurs colonnes
print(df_birth_rate)
```

```python
#Utiliser la première ligne comme en-tête
df_birth_rate
```

```python
# Supprimer le préfixe "YR" de chaque valeur dans les colonnes
df_birth_rate.columns = df_birth_rate.columns.str.replace('YR', '')
df_birth_rate
```

```python
# Statistique descriptive :
df_birth_rate.describe(include='all')
```

```python
# Supposons que votre DataFrame s'appelle df_birth_rate
# Renommer la première colonne en "Nouvelle_Colonne"
df_birth_rate.rename(columns={df_birth_rate.columns[0]: 'Code_country'}, inplace=True) # Le paramètre inplace=True applique la modification directement sur le DataFrame existant.

# Afficher le DataFrame après la modification
df_birth_rate
```

```python
df_birth_rate.columns
```

```python
#Transposer les colonnes et les lignes: from wide format to long format
df_birth_rate_unpivot = pd.melt(df_birth_rate, id_vars='Code_country', value_vars=['2000', '2001', '2002',"2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"])

#view updated DataFrame
print(df_birth_rate_unpivot)
```

```python
# Rename of second column
df_birth_rate_unpivot.rename(columns={df_birth_rate_unpivot.columns[1]: 'Year'}, inplace=True)
df_birth_rate_unpivot
```
