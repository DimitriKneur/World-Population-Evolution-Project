```python
import pandas as pd
```

```python
dataset_uhc_service_coverage = pd.read_csv('uhc_service_coverage.csv')
display(dataset_uhc_service_coverage)
```

```python
dataset_uhc_service_coverage_unpivot = pd.melt(
                                        dataset_uhc_service_coverage,
                                        id_vars = ["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 67"],
                                        var_name = "Year",
                                        value_name="uhc_service_coverage"    
                                                                )
display(dataset_uhc_service_coverage_unpivot)
```

```python
dataset_uhc_service_coverage_unpivot = dataset_uhc_service_coverage_unpivot.drop(['Indicator Name','Indicator Code','Unnamed: 67'],axis=1)
display(dataset_uhc_service_coverage_unpivot)
```

```python
dataset_uhc_service_coverage_unpivot.to_csv('dataset_uhc_service_coverage.csv')
```
