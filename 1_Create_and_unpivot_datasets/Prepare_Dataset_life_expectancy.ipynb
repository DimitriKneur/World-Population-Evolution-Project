```python
import pandas as pd
```

```python
dataset_life_expectancy = pd.read_csv('d:/Projets/Code/Python/Demoday/1_Create_and_join_datasets/total_Life_expectancy_at_birth.csv')
display(dataset_life_expectancy)
```

```python
dataset_life_expectancy_unpivot = pd.melt(
                                        dataset_life_expectancy,
                                        id_vars = "economy",
                                        var_name = "Year",
                                        value_name="Life_expectancy_in_years"    
                                                                )
display(dataset_life_expectancy_unpivot)
```

```python
dataset_life_expectancy_unpivot.rename(columns={"economy": "Country Code"}, inplace=True)

display(dataset_life_expectancy_unpivot)
```

```python
dataset_life_expectancy_unpivot["Year"] = dataset_life_expectancy_unpivot["Year"].apply(lambda x : x[2:])
display(dataset_life_expectancy_unpivot)
```

```python
dataset_life_expectancy_unpivot.to_csv('Dataset_life_expectancy.csv')
```
